```python
import pandas as pd
import pickle 
```

```python
weather_dict_path = "weather_dict.pkl"
```

```python
with open(weather_dict_path, "rb") as f:
    weather_dict = pickle.load(f)
```

```python
weather_dict
```

```python
weather_data = pd.DataFrame.from_dict(weather_dict, orient = "index")
```

```python
weather_data
```

```python
for key in weather_dict.keys():
    print(key)
```

```python
weather_df = pd.DataFrame(columns = [
    "index",  
    "temperatureMax",  
    "temperatureMin",
    "icon",
    "moonPhase",
    "windSpeed",
    "precipType",
    "precipProbability",
    "data"] )


for key in weather_dict.keys():
    df = pd.DataFrame.from_dict(weather_dict[key], orient = "index")
    df.reset_index()
    df["date"] = key
    weather_df = pd.concat([weather_df, df], axis = 1)
```
